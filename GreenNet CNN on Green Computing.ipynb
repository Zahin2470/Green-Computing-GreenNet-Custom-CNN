````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 2: Imports, Device Configuration & Global Reproducibility Seed
# ─────────────────────────────────────────────────────────────────────────────
import os, time, copy, struct, pickle, random, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from IPython.display import display
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.utils.prune as prune
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
from torch.quantization import quantize_dynamic

import torchvision
import torchvision.transforms as transforms
from torchvision import models
from torchvision import transforms
from torchvision.models import mobilenet_v2, resnet18, efficientnet_b0

from ptflops import get_model_complexity_info
from codecarbon import EmissionsTracker
from sklearn.metrics import (
    precision_score, roc_auc_score,
    confusion_matrix, classification_report
)

warnings.filterwarnings('ignore')

# ── Reproducibility ──────────────────────────────────────────────────────────
SEED = 42

def set_seed(seed=SEED):
    """Fix all sources of randomness for reproducible results."""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)

set_seed()

# ── Device ───────────────────────────────────────────────────────────────────
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"✅ Device       : {device}")
print(f"✅ CUDA version : {torch.version.cuda}")
print(f"✅ Torch version: {torch.__version__}")
print(f"✅ Random seed  : {SEED}")

# ── Training Hyperparameters (single source of truth) ────────────────────────
BATCH_TRAIN   = 64
BATCH_TEST    = 1000
LR            = 1e-3
WEIGHT_DECAY  = 1e-4
EPOCHS_MNIST  = 10   # Equal epochs for ALL models on MNIST (fair comparison)
EPOCHS_CIFAR  = 40   # Equal epochs for ALL models on CIFAR-10
PRUNE_AMOUNT  = 0.30  # 30% global L1 unstructured sparsity
KD_TEMP       = 2.0   # Knowledge Distillation temperature MINST(2) CIFAR-10(1)
KD_ALPHA      = 0.7   # KD soft-label weight (1-alpha = hard-label weight) MINST(0.7) CIFAR-10(0.3)

print(f"\n⚙️  Hyperparameters")
print(f"   Batch (train/test) : {BATCH_TRAIN}/{BATCH_TEST}")
print(f"   LR / Weight decay  : {LR} / {WEIGHT_DECAY}")
print(f"   Epochs MNIST/CIFAR : {EPOCHS_MNIST}/{EPOCHS_CIFAR}")
print(f"   Prune amount       : {PRUNE_AMOUNT*100:.0f}%")
print(f"   KD temperature (T) : {KD_TEMP}")
print(f"   KD alpha           : {KD_ALPHA}")
````


# <p style="background-color: #1e6501; color: gold; font-size: 90%; font-family: 'Lobster', cursive; text-align: center; border-radius: 15px 50px; border: 2px solid gold; padding: 20px 30px;">Section 2 : Datasets: Loading & Augmentation</p>

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 3: Dataset Loading — MNIST (IDX format) & CIFAR-10 (pickle batches)
# ─────────────────────────────────────────────────────────────────────────────

# ── 3a: IDX file reader for MNIST ────────────────────────────────────────────
def load_idx(filename):
    """Parse MNIST IDX binary format."""
    with open(filename, 'rb') as f:
        magic = f.read(4)
        dims  = struct.unpack('>B', magic[3:4])[0]
        shape = [struct.unpack('>I', f.read(4))[0] for _ in range(dims)]
        data  = np.frombuffer(f.read(), dtype=np.uint8).reshape(shape)
    return data

def resolve_idx_path(base_dir, name):
    direct_path = os.path.join(base_dir, name)
    nested_path = os.path.join(base_dir, name, name)
    if os.path.isfile(direct_path): return direct_path
    if os.path.isfile(nested_path): return nested_path
    raise FileNotFoundError(
        f"Could not find '{name}' as either:\n"
        f"  - {direct_path}\n"
        f"  - {nested_path}"
    )

mnist_dir = '/kaggle/input/datasets/hojjatk/mnist-dataset'

train_images = load_idx(resolve_idx_path(mnist_dir, 'train-images-idx3-ubyte'))
train_labels = load_idx(resolve_idx_path(mnist_dir, 'train-labels-idx1-ubyte'))
test_images  = load_idx(resolve_idx_path(mnist_dir, 't10k-images-idx3-ubyte'))
test_labels  = load_idx(resolve_idx_path(mnist_dir, 't10k-labels-idx1-ubyte'))

class MNISTDataset(Dataset):
    """Wraps raw MNIST arrays into a PyTorch Dataset."""
    def __init__(self, images, labels, transform=None):
        self.images    = images
        self.labels    = labels
        self.transform = transform

    def __len__(self):  return len(self.labels)

    def __getitem__(self, idx):
        img   = self.images[idx].astype(np.float32)
        label = int(self.labels[idx])
        if self.transform:
            img = self.transform(img)
        return img, label

mnist_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

mnist_train_ds = MNISTDataset(train_images, train_labels, mnist_transform)
mnist_test_ds  = MNISTDataset(test_images,  test_labels,  mnist_transform)

mnist_train_loader = DataLoader(mnist_train_ds, batch_size=BATCH_TRAIN,
                                shuffle=True,  num_workers=2, pin_memory=True)
mnist_test_loader  = DataLoader(mnist_test_ds,  batch_size=BATCH_TEST,
                                shuffle=False, num_workers=2, pin_memory=True)

print(f"✅ MNIST    — Train: {len(mnist_train_ds):,} | Test: {len(mnist_test_ds):,}")

# ── 3b: CIFAR-10 loader ───────────────────────────────────────────────────────
def unpickle(file):
    with open(file, 'rb') as fo:
        return pickle.load(fo, encoding='bytes')

cifar_dir = '/kaggle/input/datasets/pankrzysiu/cifar10-python/cifar-10-batches-py'

cifar_train_data, cifar_train_labels = [], []
for i in range(1, 6):
    b = unpickle(os.path.join(cifar_dir, f'data_batch_{i}'))
    cifar_train_data.append(b[b'data'])
    cifar_train_labels.extend(b[b'labels'])
cifar_train_data = np.vstack(cifar_train_data)

cifar_test_b      = unpickle(os.path.join(cifar_dir, 'test_batch'))
cifar_test_data   = cifar_test_b[b'data']
cifar_test_labels = cifar_test_b[b'labels']

CIFAR_MEAN = (0.4914, 0.4822, 0.4465)
CIFAR_STD  = (0.2023, 0.1994, 0.2010)

class CIFARDataset(Dataset):
    """Wraps raw CIFAR-10 arrays into a PyTorch Dataset."""
    def __init__(self, data, labels, transform=None):
        self.data      = data
        self.labels    = labels
        self.transform = transform

    def __len__(self):  return len(self.data)

    def __getitem__(self, idx):
        img   = self.data[idx].reshape(3, 32, 32).transpose(1, 2, 0).astype(np.uint8)
        label = int(self.labels[idx])
        if self.transform:
            img = self.transform(img)
        return img, label

# ── Cutout augmentation ───────────────────────────────────────────────────────
class Cutout:
    """
    Randomly masks a square patch of size×size pixels with zeros.
    Forces the model to rely on global context rather than local patches.
    Proven +2-3% accuracy on CIFAR-10 at no extra compute cost.
    Applied ONLY to training — never to test images.
    """
    def __init__(self, size=16):
        self.size = size

    def __call__(self, img):
        h, w = img.shape[1], img.shape[2]
        cx   = np.random.randint(w)
        cy   = np.random.randint(h)
        x1   = max(0, cx - self.size // 2)
        x2   = min(w, cx + self.size // 2)
        y1   = max(0, cy - self.size // 2)
        y2   = min(h, cy + self.size // 2)
        img[:, y1:y2, x1:x2] = 0        # zero out the patch
        return img

# ── CIFAR-10 transforms ───────────────────────────────────────────────────────
cifar_train_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize(CIFAR_MEAN, CIFAR_STD),
    Cutout(size=16),                     # ← after ToTensor (operates on tensor)
])

cifar_test_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.ToTensor(),
    transforms.Normalize(CIFAR_MEAN, CIFAR_STD)
    # NO Cutout on test — evaluation must be clean
])

cifar_train_ds = CIFARDataset(cifar_train_data,  cifar_train_labels, cifar_train_transform)
cifar_test_ds  = CIFARDataset(cifar_test_data,   cifar_test_labels,  cifar_test_transform)

cifar_train_loader = DataLoader(cifar_train_ds, batch_size=BATCH_TRAIN,
                                shuffle=True,  num_workers=2, pin_memory=True)
cifar_test_loader  = DataLoader(cifar_test_ds,  batch_size=BATCH_TEST,
                                shuffle=False, num_workers=2, pin_memory=True)

print(f"✅ CIFAR-10 — Train: {len(cifar_train_ds):,} | Test: {len(cifar_test_ds):,}")
print(f"Augmentation: RandomCrop + HFlip + ColorJitter + Cutout(16)")

CIFAR_CLASSES = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
````


# <p style="background-color: #1e6501; color: gold; font-size: 90%; font-family: 'Lobster', cursive; text-align: center; border-radius: 15px 50px; border: 2px solid gold; padding: 20px 30px;">Section 3 : CNN Model Architectures</p>

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 5: GreenNet — Proposed Lightweight Architecture
#
#   • Depthwise-Separable Convolutions (DSC) to significantly reduce parameters and MACs
#   • Residual connections (ResidualDSBlock) to improve gradient flow and stabilize deeper networks
#   • Batch Normalization after each convolution for faster and stable training
#   • ReLU6 activation for quantization-friendly bounded non-linearity
#   • Progressive multi-stage feature extraction (4 stages for CIFAR-10)
#   • Global Average Pooling (GAP) to eliminate large fully connected layers
#   • Lightweight classifier head with reduced Dropout (0.2) to balance regularization
#
# Architecture highlights:
#   • Stage 1: Standard Conv + BN + ReLU6 + MaxPool (initial feature extraction)
#   • Stage 2–4: Residual Depthwise-Separable Blocks (efficient + deeper representation)
#   • Channel scaling controlled via base_channels: MNIST  : 16 → 32 → 48 → 64  CIFAR-10: 64 → 128 → 192 → 256
#
# Efficiency:
#   • DSC reduces computation compared to standard convolutions
#   • Residual design enables deeper network without degradation
#   • Model remains compact (~16K params for MNIST, ~520K for CIFAR)
# ─────────────────────────────────────────────────────────────────────────────

class DepthwiseSeparableBlock(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.dw    = nn.Conv2d(in_ch, in_ch,  kernel_size=3, padding=1, groups=in_ch, bias=False)
        self.pw    = nn.Conv2d(in_ch, out_ch, kernel_size=1, bias=False)
        self.bn_dw = nn.BatchNorm2d(in_ch)
        self.bn_pw = nn.BatchNorm2d(out_ch)
        self.act   = nn.ReLU6(inplace=True)

    def forward(self, x):
        x = self.act(self.bn_dw(self.dw(x)))
        x = self.act(self.bn_pw(self.pw(x)))
        return x


class ResidualDSBlock(nn.Module):
    """
    DSC block with residual skip connection.
    Skip uses 1×1 conv to match channels when in_ch != out_ch.
    Residuals dramatically improve gradient flow — critical for 85%+ CIFAR.
    """
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.dsblock = DepthwiseSeparableBlock(in_ch, out_ch)
        # 1×1 projection for skip when channel dims differ
        self.skip = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, kernel_size=1, bias=False),
            nn.BatchNorm2d(out_ch)
        ) if in_ch != out_ch else nn.Identity()
        self.act = nn.ReLU6(inplace=True)

    def forward(self, x):
        return self.act(self.dsblock(x) + self.skip(x))


class GreenNet(nn.Module):
    """
    Redesigned for CIFAR-10 85%+ target.

    MNIST  (base_channels=16): 16→32→48          ~12K params  <0.1MB
    CIFAR  (base_channels=64): 64→128→192→256    ~174K params ~1MB

    Key changes:
      • ResidualDSBlock — skip connections fix vanishing gradients in 4-stage net
      • 4th stage added — deeper feature hierarchy needed for natural images
      • base_channels=64 for CIFAR — 4× wider than before (was 32→64→96)
      • Dropout reduced 0.2 — wider model less prone to overfitting
    """
    def __init__(self, num_classes=10, in_channels=1, base_channels=16):
        super().__init__()
        c1 = base_channels          # 16  / 64
        c2 = base_channels * 2      # 32  / 128
        c3 = base_channels * 3      # 48  / 192
        c4 = base_channels * 4      # 64  / 256

        # Stage 1 — standard entry conv
        self.stage1 = nn.Sequential(
            nn.Conv2d(in_channels, c1, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(c1),
            nn.ReLU6(inplace=True),
            nn.MaxPool2d(2, 2)
        )
        # Stage 2 — residual DSC
        self.stage2 = nn.Sequential(
            ResidualDSBlock(c1, c2),
            nn.MaxPool2d(2, 2)
        )
        # Stage 3 — residual DSC
        self.stage3 = ResidualDSBlock(c2, c3)

        # Stage 4 — residual DSC (new — deeper hierarchy)
        self.stage4 = ResidualDSBlock(c3, c4)

        self.gap  = nn.AdaptiveAvgPool2d(1)
        self.drop = nn.Dropout(0.2)
        self.fc   = nn.Linear(c4, num_classes)

    def forward(self, x):
        x = self.stage1(x)
        x = self.stage2(x)
        x = self.stage3(x)
        x = self.stage4(x)
        x = self.gap(x)
        x = x.view(x.size(0), -1)
        x = self.drop(x)
        return self.fc(x)


set_seed()
custom_mnist = GreenNet(num_classes=10, in_channels=1, base_channels=16).to(device)
custom_cifar = GreenNet(num_classes=10, in_channels=3, base_channels=64).to(device)

for name, m in [('MNIST  (base_channels=16)', custom_mnist),
                ('CIFAR  (base_channels=64)', custom_cifar)]:
    p = sum(x.numel() for x in m.parameters())
    from ptflops import get_model_complexity_info
    macs, _ = get_model_complexity_info(m.cpu(), (1,28,28) if 'MNIST' in name else (3,32,32),
                                         as_strings=False, print_per_layer_stat=False, verbose=False)
    m.to(device)
    print(f"✅ GreenNet {name}: {p:,} params | {macs/1e9:.4f} GMACs")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 7: Metric Utilities
# (Giga Multiply-Accumulate Operations). 1 GMAC ≈ 2 GFLOPs.
# ─────────────────────────────────────────────────────────────────────────────

def count_params(model):
    """Count trainable parameters."""
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

def get_gmacs(model, input_shape):
    """
    Compute GMACs safely. Returns None if ptflops fails
    (e.g. quantized models whose ops are incompatible with ptflops).
    """
    try:
        macs, _ = get_model_complexity_info(
            model, input_shape,
            as_strings=False,
            print_per_layer_stat=False,
            verbose=False
        )
        return macs / 1e9 if macs is not None else None
    except Exception:
        return None   # caller must handle None for quantized models

def get_model_size_mb(model):
    """Model size on disk (MB)."""
    path = '/tmp/_tmp_model.pt'
    torch.save(model.state_dict(), path)
    size = os.path.getsize(path) / 1e6
    os.remove(path)
    return size

def get_inference_latency_ms(model, loader, device):
    """
    Average per-image inference latency (ms).
    Runs over the full test set to reduce timing noise.
    """
    model.eval()
    # GPU warm-up
    if device.type == 'cuda':
        dummy = next(iter(loader))[0][:2].to(device)
        with torch.no_grad():
            _ = model(dummy)
    start = time.perf_counter()
    total = 0
    with torch.no_grad():
        for inputs, _ in loader:
            _ = model(inputs.to(device))
            total += inputs.size(0)
    elapsed = time.perf_counter() - start
    return (elapsed / total) * 1000  # ms per image

def evaluate_model(model, loader, device):
    """
    Full evaluation: accuracy, macro-precision, loss, macro-ROC-AUC.
    Returns (accuracy, precision, loss, roc_auc, all_preds, all_labels).
    """
    model.eval()
    criterion  = nn.CrossEntropyLoss()
    total_loss = correct = total = 0
    all_preds, all_labels, all_probs = [], [], []

    with torch.no_grad():
        for inputs, labels in loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            total_loss += criterion(outputs, labels).item()
            preds = outputs.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total   += labels.size(0)
            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())
            all_probs.extend(torch.softmax(outputs, 1).cpu().tolist())

    acc  = correct / total
    prec = precision_score(all_labels, all_preds, average='macro', zero_division=0)
    loss = total_loss / len(loader)
    roc  = roc_auc_score(all_labels, np.array(all_probs), multi_class='ovr',
                         average='macro') if len(set(all_labels)) > 1 else 0.0
    return acc, prec, loss, roc, all_preds, all_labels

def compute_green_score(accuracy, co2_kg, size_mb, latency_ms):
    """
    GreenScore = Accuracy / (CO₂_kg × Size_MB × Latency_ms)^(1/3)
    Note: CO₂ in kg, size in MB, latency in ms/image.
    """
    denom = (max(co2_kg, 1e-12) * max(size_mb, 1e-6) * max(latency_ms, 1e-6)) ** (1/3)
    return accuracy / denom

print("✅ Metric utilities defined.")
print("Note: Computational cost reported as GMACs (Giga Multiply-Accumulate Ops)")
print("1 GMAC ≈ 2 GFLOPs. GMACs are the standard in efficient-CNN literature.")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 6: Pretrained Baseline Models
# MobileNetV2, ResNet18, EfficientNet-B0 — adapted for 1-ch MNIST & 3-ch CIFAR
# ─────────────────────────────────────────────────────────────────────────────

def adapt_mobilenetv2(in_channels=1, num_classes=10):
    """MobileNetV2 adapted for arbitrary input channels."""
    m = mobilenet_v2(pretrained=True)
    if in_channels != 3:
        old = m.features[0][0]  # first conv
        m.features[0][0] = nn.Conv2d(
            in_channels, old.out_channels,
            kernel_size=old.kernel_size, stride=old.stride,
            padding=old.padding, bias=False
        )
    m.classifier[1] = nn.Linear(m.last_channel, num_classes)
    return m.to(device)

def adapt_resnet18(in_channels=1, num_classes=10):
    """ResNet18 adapted for arbitrary input channels."""
    m = resnet18(pretrained=True)
    if in_channels != 3:
        old = m.conv1
        m.conv1 = nn.Conv2d(
            in_channels, old.out_channels,
            kernel_size=old.kernel_size, stride=old.stride,
            padding=old.padding, bias=False
        )
    m.fc = nn.Linear(m.fc.in_features, num_classes)
    return m.to(device)

def adapt_efficientnet_b0(in_channels=1, num_classes=10):
    """EfficientNet-B0 adapted for arbitrary input channels."""
    m = efficientnet_b0(pretrained=True)
    if in_channels != 3:
        old = m.features[0][0]
        m.features[0][0] = nn.Conv2d(
            in_channels, old.out_channels,
            kernel_size=old.kernel_size, stride=old.stride,
            padding=old.padding, bias=False
        )
    m.classifier[1] = nn.Linear(m.classifier[1].in_features, num_classes)
    return m.to(device)

set_seed()
mobilenet_mnist    = adapt_mobilenetv2(in_channels=1)
mobilenet_cifar    = adapt_mobilenetv2(in_channels=3)
resnet_mnist       = adapt_resnet18(in_channels=1)
resnet_cifar       = adapt_resnet18(in_channels=3)
efficientnet_mnist = adapt_efficientnet_b0(in_channels=1)
efficientnet_cifar = adapt_efficientnet_b0(in_channels=3)

print("✅ Pretrained baselines loaded:")
for name, m in [('MobileNetV2',   mobilenet_mnist),
                ('ResNet18',       resnet_mnist),
                ('EfficientNet-B0',efficientnet_mnist)]:
    p = sum(x.numel() for x in m.parameters())
    print(f"   {name:20s}: {p:>12,} parameters")
````

# <p style="background-color: #1e6501; color: gold; font-size: 90%; font-family: 'Lobster', cursive; text-align: center; border-radius: 15px 50px; border: 2px solid gold; padding: 20px 30px;">Section 4 : Training & Evaluation Infrastructure</p>

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 8: Training Function
#
# Why: label_smoothing=0.1 was penalizing confident correct predictions.
# For a tiny 189K-parameter model on a clean dataset (MNIST has no label noise), added artificial uncertainty the model could not overcome.
#
# Rule of thumb:
#   • label_smoothing > 0  → only helps when labels are noisy or ambiguous (e.g., CIFAR-100, real-world web-scraped data)
#   • label_smoothing = 0  → correct choice for clean, unambiguous datasets like MNIST with a low-capacity model
#
# Callers that need smoothing (e.g., CIFAR-10 baselines) can still pass
# label_smoothing explicitly. Default is now 0.0 (safe for all cases).
# ─────────────────────────────────────────────────────────────────────────────

def train_model(model, train_loader, test_loader, epochs, name, device,
                label_smoothing=0.0,
                verbose=True):
    criterion = nn.CrossEntropyLoss(label_smoothing=label_smoothing)
    optimizer = optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-6)

    tracker = EmissionsTracker(
        project_name=name,
        log_level='error',
        save_to_file=False,
        measure_power_secs=5
    )
    tracker.start()

    history = {'train_loss': [], 'val_acc': []}
    best_val_acc = 0.0

    for epoch in range(1, epochs + 1):
        model.train()
        running_loss = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad(set_to_none=True)
            loss = criterion(model(inputs), labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        scheduler.step()

        acc, _, _, _, _, _ = evaluate_model(model, test_loader, device)
        avg_loss = running_loss / len(train_loader)
        history['train_loss'].append(avg_loss)
        history['val_acc'].append(acc)
        if acc > best_val_acc:
            best_val_acc = acc

        if verbose and (epoch == 1 or epoch % 5 == 0 or epoch == epochs):
            lr_now = scheduler.get_last_lr()[0]
            print(f"   Epoch {epoch:3d}/{epochs} | Loss: {avg_loss:.4f} | "
                  f"Val Acc: {acc:.4f} | LR: {lr_now:.2e}")

    co2_kg = tracker.stop()
    if verbose:
        print(f"   ✅ {name} — Best val acc: {best_val_acc:.4f} | CO₂: {co2_kg:.6f} kg")
    return co2_kg, history
````

# <p style="background-color: #1e6501; color: gold; font-size: 90%; font-family: 'Lobster', cursive; text-align: center; border-radius: 15px 50px; border: 2px solid gold; padding: 20px 30px;">Section 4.1 : Compression Utilities — Pruning & Quantization</p>

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 9: Compression Utilities — Pruning & Quantization
# ─────────────────────────────────────────────────────────────────────────────

def apply_global_pruning(model, amount=PRUNE_AMOUNT, fine_tune_loader=None,
                          fine_tune_epochs=3, device=device):
    """
    Global L1 unstructured pruning + optional fine-tuning to recover accuracy.

    v1 bug: pruning was applied but NOT fine-tuned, causing accuracy degradation.
    This version fine-tunes the pruned model for recovery.
    """
    model = copy.deepcopy(model)
    targets = [(m, 'weight') for m in model.modules()
               if isinstance(m, (nn.Conv2d, nn.Linear))]

    prune.global_unstructured(
        targets,
        pruning_method=prune.L1Unstructured,
        amount=amount
    )

    # Make pruning permanent (actual weight removal for real speedup)
    for m, name in targets:
        prune.remove(m, name)

    # ✅ Fine-tune to recover accuracy (critical step missing in v1)
    if fine_tune_loader is not None and fine_tune_epochs > 0:
        ft_criterion = nn.CrossEntropyLoss()
        ft_optimizer = optim.Adam(model.parameters(), lr=LR * 0.1, weight_decay=WEIGHT_DECAY)
        model.train()
        for _ in range(fine_tune_epochs):
            for inputs, labels in fine_tune_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                ft_optimizer.zero_grad(set_to_none=True)
                ft_criterion(model(inputs), labels).backward()
                ft_optimizer.step()

    sparsity = sum((p == 0).sum().item() for p in model.parameters()) / \
               sum(p.numel() for p in model.parameters())
    print(f"   Pruned at {amount*100:.0f}% | Actual sparsity: {sparsity*100:.1f}%"
          + (f" | Fine-tuned {fine_tune_epochs} epochs" if fine_tune_loader else ""))
    return model


def apply_dynamic_quantization(model):
    """
    8-bit dynamic quantization (INT8).
    Quantizes Linear layers (full CUDA support).
    Returns CPU model (dynamic quant is CPU-optimized).
    """
    model_cpu = copy.deepcopy(model).cpu()
    quantized  = quantize_dynamic(
        model_cpu,
        {nn.Linear},
        dtype=torch.qint8
    )
    return quantized  # stays on CPU — dynamic quant is CPU-only


print("✅ Pruning and quantization utilities defined.")
````

# <p style="background-color: #1e6501; color: gold; font-size: 90%; font-family: 'Lobster', cursive; text-align: center; border-radius: 15px 50px; border: 2px solid gold; padding: 20px 30px;">Section 4.2 : Knowledge Distillation</p>

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 10: Knowledge Distillation — Novel Contribution
#
# We train GreenNet (student) to mimic MobileNetV2 (teacher).
# Loss = α * KL(soft teacher | soft student) + (1-α) * CE(hard labels | student)
#
# KD allows the small model to learn rich feature representations from
# a larger pre-trained teacher without adding any inference cost.
# ─────────────────────────────────────────────────────────────────────────────

def train_with_distillation(student, teacher, train_loader, test_loader,
                            epochs, name, device,
                            temperature=KD_TEMP, alpha=KD_ALPHA):
    """
    Knowledge Distillation training.

    Parameters
    ----------
    student     : lightweight model being trained
    teacher     : frozen pretrained model providing soft targets
    temperature : T — softens probability distributions (higher T → softer)
    alpha       : weight for soft (KD) loss; (1-alpha) for hard (CE) loss

    Returns
    -------
    co2_kg  : training CO₂ emissions in kg
    history : per-epoch metrics
    """
    teacher.eval()   # Teacher is always in eval mode — frozen
    for p in teacher.parameters():
        p.requires_grad_(False)

    criterion_ce = nn.CrossEntropyLoss(label_smoothing=0.05)
    criterion_kd = nn.KLDivLoss(reduction='batchmean')
    optimizer    = optim.Adam(student.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
    scheduler    = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-6)

    tracker = EmissionsTracker(project_name=f"KD_{name}", log_level='error', save_to_file=False)
    tracker.start()

    history = {'train_loss': [], 'val_acc': []}
    best_val_acc = 0.0

    for epoch in range(1, epochs + 1):
        student.train()
        running_loss = 0.0

        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            with torch.no_grad():
                teacher_logits = teacher(inputs)

            student_logits = student(inputs)

            # Soft targets (KD loss)
            soft_student = F.log_softmax(student_logits / temperature, dim=1)
            soft_teacher = F.softmax(teacher_logits  / temperature, dim=1)
            loss_kd = criterion_kd(soft_student, soft_teacher) * (temperature ** 2)

            # Hard labels (CE loss)
            loss_ce = criterion_ce(student_logits, labels)

            # Combined loss
            loss = alpha * loss_kd + (1 - alpha) * loss_ce

            optimizer.zero_grad(set_to_none=True)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        scheduler.step()
        acc, _, _, _, _, _ = evaluate_model(student, test_loader, device)
        avg_loss = running_loss / len(train_loader)
        history['train_loss'].append(avg_loss)
        history['val_acc'].append(acc)
        if acc > best_val_acc:
            best_val_acc = acc

        if epoch == 1 or epoch % 5 == 0 or epoch == epochs:
            print(f"   [KD] Epoch {epoch:3d}/{epochs} | Loss: {avg_loss:.4f} | "
                  f"Val Acc: {acc:.4f}")

    co2_kg = tracker.stop()
    print(f"   ✅ KD {name} — Best val acc: {best_val_acc:.4f} | CO₂: {co2_kg:.6f} kg")
    return co2_kg, history

print("✅ Knowledge Distillation trainer defined.")
````

# <p style="background-color: #1e6501; color: gold; font-size: 90%; font-family: 'Lobster', cursive; text-align: center; border-radius: 15px 50px; border: 2px solid gold; padding: 20px 30px;">Section 5 : MNIST Benchmark</p>

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 11: MNIST — Training All Models (Equal epochs for fair comparison)
# ─────────────────────────────────────────────────────────────────────────────
MNIST_INPUT_SHAPE = (1, 28, 28)

def train_model_mnist_custom(model, train_loader, test_loader, epochs, name, device):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(
        model.parameters(), lr=0.05,
        momentum=0.9, weight_decay=WEIGHT_DECAY, nesterov=True
    )
    WARMUP = 3
    def lr_lambda(ep):
        if ep < WARMUP:
            return (ep + 1) / WARMUP
        progress = (ep - WARMUP) / max(1, epochs - WARMUP)
        return 0.5 * (1.0 + np.cos(np.pi * progress))
    scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

    tracker = EmissionsTracker(project_name=name, log_level='error',
                               save_to_file=False, measure_power_secs=5)
    tracker.start()
    history = {'train_loss': [], 'val_acc': []}
    best_val_acc = 0.0

    for epoch in range(1, epochs + 1):
        model.train()
        running_loss = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad(set_to_none=True)
            loss = criterion(model(inputs), labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        scheduler.step()
        acc, _, _, _, _, _ = evaluate_model(model, test_loader, device)
        avg_loss = running_loss / len(train_loader)
        history['train_loss'].append(avg_loss)
        history['val_acc'].append(acc)
        if acc > best_val_acc:
            best_val_acc = acc
        if epoch == 1 or epoch % 2 == 0 or epoch == epochs:
            print(f"   Epoch {epoch:3d}/{epochs} | Loss: {avg_loss:.4f} | "
                  f"Val Acc: {acc:.4f} | LR: {scheduler.get_last_lr()[0]:.2e}")

    co2_kg = tracker.stop()
    print(f"   ✅ {name} — Best val acc: {best_val_acc:.4f} | CO₂: {co2_kg:.6f} kg")
    return co2_kg, history

print(f"{'='*70}")
print(f" MNIST BENCHMARK — FAIR COMPARISON")
print(f" All Models Epochs: {EPOCHS_MNIST} epochs")
print(f"{'='*70}")

# ── 1. GreenNet ──────────────────────────────────────────────────────────
print("\n[1/6] GreenNet (SGD + Nesterov, warmup, no label smoothing)")
set_seed()
custom_mnist = GreenNet(in_channels=1).to(device)
co2_custom_mnist, hist_custom_mnist = train_model_mnist_custom(
    custom_mnist, mnist_train_loader, mnist_test_loader, epochs=10, name='GreenNet_MNIST', device=device)

# ── 2. GreenNet + KD ─────────────────────────────────────────────────────
print("\n[2/6] GreenNet + Knowledge Distillation (teacher: MobileNetV2)")
set_seed()
custom_mnist_kd = GreenNet(in_channels=1).to(device)

# Warm up teacher for 3 epochs (was 2) so soft labels are meaningful
_, _ = train_model(
    mobilenet_mnist, mnist_train_loader, mnist_test_loader, 3, 'Teacher_warmup_MNIST', device,
    label_smoothing=0.0, verbose=False)

co2_kd_mnist, hist_kd_mnist = train_with_distillation(
    custom_mnist_kd, mobilenet_mnist,
    mnist_train_loader, mnist_test_loader,
    epochs=10, name='MNIST', device=device,
    temperature=2,    # 2 — T=2 over-smooths MNIST's sharp class distributions
    alpha=0.7)        # 0.7 — let CE loss dominate more (0.3 weight) over KD (0.7)

# ── 3. MobileNetV2 ────────────────────────────────────────────────────────────
print(f"\n[3/6] MobileNetV2 (full pretrained) — {EPOCHS_MNIST} epochs")
set_seed()
mobilenet_mnist = adapt_mobilenetv2(in_channels=1)
co2_mob_mnist, hist_mob_mnist = train_model(
    mobilenet_mnist, mnist_train_loader, mnist_test_loader,
    EPOCHS_MNIST, 'MobileNetV2_MNIST', device)

# ── 4. MobileNetV2 Pruned 30% ─────────────────────────────────────────────────
print("\n[4/6] MobileNetV2 Pruned 30% + Fine-tuned")
pruned_mob_mnist = apply_global_pruning(
    mobilenet_mnist, PRUNE_AMOUNT, mnist_train_loader, fine_tune_epochs=3)

# ── 5. ResNet18 Pruned 30% ────────────────────────────────────────────────────
print(f"\n[5/6] ResNet18 Pruned 30% — {EPOCHS_MNIST} epochs")
set_seed()
resnet_mnist = adapt_resnet18(in_channels=1)
co2_res_mnist, _ = train_model(
    resnet_mnist, mnist_train_loader, mnist_test_loader,
    EPOCHS_MNIST, 'ResNet18_MNIST', device)
pruned_res_mnist = apply_global_pruning(
    resnet_mnist, PRUNE_AMOUNT, mnist_train_loader, fine_tune_epochs=3)

# ── 6. EfficientNet-B0 Pruned 30% ─────────────────────────────────────────────
print(f"\n[6/6] EfficientNet-B0 Pruned 30% — {EPOCHS_MNIST} epochs")
set_seed()
efficientnet_mnist = adapt_efficientnet_b0(in_channels=1)
co2_eff_mnist, _ = train_model(
    efficientnet_mnist, mnist_train_loader, mnist_test_loader,
    EPOCHS_MNIST, 'EfficientNetB0_MNIST', device)
pruned_eff_mnist = apply_global_pruning(
    efficientnet_mnist, PRUNE_AMOUNT, mnist_train_loader, fine_tune_epochs=3)

print("\n✅ All MNIST models trained.")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 13: MNIST — Collect All Metrics
# Pruned models use base training CO₂ (fine-tune CO₂ not tracked)
# fallback_gmacs added for INT8 models
# ─────────────────────────────────────────────────────────────────────────────

def collect_metrics(model, loader, input_shape, co2_kg, name, device,
                    is_cpu=False, fallback_gmacs=None):
    eval_device = torch.device('cpu') if is_cpu else device

    # GMACs — must run on CPU (ptflops requirement)
    if fallback_gmacs is not None:
        gmacs = fallback_gmacs
    else:
        model.cpu()
        gmacs = get_gmacs(model, input_shape)
        if gmacs is None:
            gmacs = 0.0
    params = count_params(model)

    # Move to eval_device BEFORE any inference
    model.to(eval_device)
    acc, prec, loss, roc, preds, labels = evaluate_model(model, loader, eval_device)
    size_mb = get_model_size_mb(model)
    lat_ms  = get_inference_latency_ms(model, loader, eval_device)
    gs      = compute_green_score(acc, co2_kg, size_mb, lat_ms)

    # Return model to main device
    model.to(device)

    return {
        'Model':        name,
        'Accuracy':     acc,
        'Precision':    prec,
        'Loss':         loss,
        'ROC-AUC':      roc,
        'Params':       params,
        'GMACs':        gmacs,
        'Size (MB)':    size_mb,
        'Latency (ms)': lat_ms,
        'CO₂ (kg)':     co2_kg,
        'Green Score':  gs,
        '_preds':       preds,
        '_labels':      labels
    }

print("Collecting MNIST metrics...")
mnist_results = []

mnist_results.append(collect_metrics(
    custom_mnist, mnist_test_loader, MNIST_INPUT_SHAPE,
    co2_custom_mnist, 
    'GreenNet', device))

mnist_results.append(collect_metrics(
    custom_mnist_kd, mnist_test_loader, MNIST_INPUT_SHAPE,
    co2_kd_mnist,  
    'GreenNet + KD', device))

mnist_results.append(collect_metrics(
    mobilenet_mnist, mnist_test_loader, MNIST_INPUT_SHAPE,
    co2_mob_mnist,      
    'MobileNetV2 (Full)', device))

mnist_results.append(collect_metrics(
    pruned_mob_mnist, mnist_test_loader, MNIST_INPUT_SHAPE,
    co2_mob_mnist,              # ← base training CO₂ (pruning fine-tune not tracked separately)
    'MobileNetV2 Pruned 30%', device))

mnist_results.append(collect_metrics(
    pruned_res_mnist, mnist_test_loader, MNIST_INPUT_SHAPE,
    co2_res_mnist,         
    'ResNet18 Pruned 30%', device))

mnist_results.append(collect_metrics(
    pruned_eff_mnist, mnist_test_loader, MNIST_INPUT_SHAPE,
    co2_eff_mnist,       
    'EfficientNet-B0 Pruned 30%', device))

# ── Quantized models — reuse base GMACs ───────────────
_custom_gmacs = mnist_results[0]['GMACs']   # GreenNet base GMACs
_mob_gmacs    = mnist_results[2]['GMACs']   # MobileNetV2 base GMACs

mnist_results.append(collect_metrics(
    quant_custom_mnist, mnist_test_loader, MNIST_INPUT_SHAPE,
    co2_quant_custom_mnist, 
    'GreenNet INT8', device,
    is_cpu=True, fallback_gmacs=_custom_gmacs))

mnist_results.append(collect_metrics(
    quant_mob_mnist, mnist_test_loader, MNIST_INPUT_SHAPE,
    co2_quant_mob_mnist,        # ← real value from Cell 12 (= co2_mob_mnist)
    'MobileNetV2 INT8', device,
    is_cpu=True, fallback_gmacs=_mob_gmacs))

mnist_df = pd.DataFrame([{k: v for k, v in r.items() if not k.startswith('_')}
                          for r in mnist_results])

print("\n" + "="*90)
print(f"{'MNIST - GREEN AI BENCHMARK':^70}")
print(f"{'All models: ' + str(EPOCHS_MNIST) + ' epochs':^90}")
print("="*90)

display(mnist_df.style
    .background_gradient(cmap='RdYlGn',   subset=['Accuracy', 'Green Score'])
    .background_gradient(cmap='RdYlGn_r', subset=['CO₂ (kg)', 'Size (MB)', 'GMACs'])
    .format({
        'Accuracy':     '{:.4f}',
        'Precision':    '{:.4f}',
        'Loss':         '{:.4f}',
        'ROC-AUC':      '{:.4f}',
        'Params':       '{:,}',
        'GMACs':        '{:.4f}',
        'Size (MB)':    '{:.2f}',
        'Latency (ms)': '{:.3f}',
        'CO₂ (kg)':     '{:.6f}',
        'Green Score':  '{:.1f}',
    })
)
print("="*90)
````

````python
# Cell 14: Specialized trainer for GreenNet on CIFAR-10 ────────────────────────
#   • label_smoothing=0.01 — CIFAR-10 has genuine inter-class ambiguity
#   • 40 epochs — compact model needs more steps to converge on natural images
#   • Warmup 8 epochs then cosine — prevents early LR collapse

CIFAR_INPUT_SHAPE = (3, 32, 32)

def train_model_cifar_custom(model, train_loader, test_loader, epochs, name, device):
    """
    Key changes vs previous:
      • OneCycleLR — best-in-class single-cycle schedule for CIFAR,
        peaks at max_lr=0.05 then anneals with cosine + momentum cycling
      • epochs=0 — wider/deeper model needs full convergence time
      • label_smoothing=0.1 — stronger regularization for wider model
        that could otherwise overfit CIFAR-10 training set
    """
    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
    optimizer = optim.SGD(
        model.parameters(),
        lr=0.01,
        momentum=0.9,
        weight_decay=5e-4,          # stronger WD for wider model
        nesterov=True
    )
    # OneCycleLR: ramps up to max_lr then decays — proven best schedule for CIFAR
    scheduler = optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=0.05,                # peak LR after warmup
        steps_per_epoch=len(train_loader),
        epochs=epochs,
        pct_start=0.3,              # 30% warmup, 70% decay
        anneal_strategy='cos',
        div_factor=5.0,             # initial_lr = max_lr/5 = 0.01
        final_div_factor=1e4        # final_lr = initial_lr/1e4
    )

    tracker = EmissionsTracker(project_name=name, log_level='error',
                               save_to_file=False, measure_power_secs=5)
    tracker.start()
    history      = {'train_loss': [], 'val_acc': []}
    best_val_acc = 0.0

    for epoch in range(1, epochs + 1):
        model.train()
        running_loss = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad(set_to_none=True)
            loss = criterion(model(inputs), labels)
            loss.backward()
            # Gradient clipping — prevents occasional large steps with OneCycle
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
            scheduler.step()        # OneCycleLR steps per BATCH not per epoch
            running_loss += loss.item()

        acc, _, _, _, _, _ = evaluate_model(model, test_loader, device)
        avg_loss = running_loss / len(train_loader)
        history['train_loss'].append(avg_loss)
        history['val_acc'].append(acc)
        if acc > best_val_acc:
            best_val_acc = acc
        if epoch == 1 or epoch % 5 == 0 or epoch == epochs:
            print(f"   Epoch {epoch:3d}/{epochs} | Loss: {avg_loss:.4f} | "
                  f"Val Acc: {acc:.4f} | LR: {scheduler.get_last_lr()[0]:.2e}")

    co2_kg = tracker.stop()
    print(f"   ✅ {name} — Best val acc: {best_val_acc:.4f} | CO₂: {co2_kg:.6f} kg")
    return co2_kg, history


print(f"{'='*70}")
print(f" CIFAR-10 BENCHMARK — FAIR COMPARISON")
print(f" GreenNet and All Models: {EPOCHS_CIFAR} epochs")
print(f"{'='*70}")

# ── 1. GreenNet ──────────────────────────────────────────────────────────
print("\n[1/6] GreenNet (base_channels=64, residual, OneCycleLR, 40 epochs)")
set_seed()
custom_cifar = GreenNet(in_channels=3, base_channels=64).to(device)
co2_custom_cifar, hist_custom_cifar = train_model_cifar_custom(
    custom_cifar, cifar_train_loader, cifar_test_loader, epochs=40, name='GreenNet_CIFAR', device=device)

# ── 2. GreenNet + KD ─────────────────────────────────────────────────────
print("\n[2/6] GreenNet + Knowledge Distillation (teacher: MobileNetV2)")
set_seed()
custom_cifar_kd = GreenNet(in_channels=3, base_channels=64).to(device)

# 12 teacher warmup epochs — fully converged teacher = much richer soft labels
print("   Warming up teacher (5 epochs)...")
_, _ = train_model(
    mobilenet_cifar, cifar_train_loader, cifar_test_loader, 5, 'Teacher_warmup_CIFAR', device,
    label_smoothing=0.0, verbose=False)
acc_teacher, _, _, _, _, _ = evaluate_model(mobilenet_cifar, cifar_test_loader, device)
print(f"   Teacher accuracy after warmup: {acc_teacher:.4f}")

co2_kd_cifar, hist_kd_cifar = train_with_distillation(
    custom_cifar_kd, mobilenet_cifar,
    cifar_train_loader, cifar_test_loader,
    epochs=40, name='CIFAR', device=device,
    temperature=1,
    alpha=0.3)

# ── 3. MobileNetV2 ────────────────────────────────────────────────────────────
print(f"\n[3/6] MobileNetV2 (full pretrained) — {EPOCHS_CIFAR} epochs")
set_seed()
mobilenet_cifar = adapt_mobilenetv2(in_channels=3)
co2_mob_cifar, hist_mob_cifar = train_model(
    mobilenet_cifar, cifar_train_loader, cifar_test_loader,
    EPOCHS_CIFAR, 'MobileNetV2_CIFAR', device)

# ── 4. MobileNetV2 Pruned 30% ─────────────────────────────────────────────────
print("\n[4/6] MobileNetV2 Pruned 30% + Fine-tuned")
pruned_mob_cifar = apply_global_pruning(
    mobilenet_cifar, PRUNE_AMOUNT, cifar_train_loader, fine_tune_epochs=3)

# ── 5. ResNet18 Pruned 30% ────────────────────────────────────────────────────
print(f"\n[5/6] ResNet18 Pruned 30% — {EPOCHS_CIFAR} epochs")
set_seed()
resnet_cifar = adapt_resnet18(in_channels=3)
co2_res_cifar, _ = train_model(
    resnet_cifar, cifar_train_loader, cifar_test_loader,
    EPOCHS_CIFAR, 'ResNet18_CIFAR', device)
pruned_res_cifar = apply_global_pruning(
    resnet_cifar, PRUNE_AMOUNT, cifar_train_loader, fine_tune_epochs=3)

# ── 6. EfficientNet-B0 Pruned 30% ─────────────────────────────────────────────
print(f"\n[6/6] EfficientNet-B0 Pruned 30% — {EPOCHS_CIFAR} epochs")
set_seed()
efficientnet_cifar = adapt_efficientnet_b0(in_channels=3)
co2_eff_cifar, _ = train_model(
    efficientnet_cifar, cifar_train_loader, cifar_test_loader,
    EPOCHS_CIFAR, 'EfficientNetB0_CIFAR', device)
pruned_eff_cifar = apply_global_pruning(
    efficientnet_cifar, PRUNE_AMOUNT, cifar_train_loader, fine_tune_epochs=3)

print("\n✅ All CIFAR-10 models trained.")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 16: Progressive Pruning Sweep — Accuracy vs Sparsity
# GreenNet gets 2-epoch fine-tune per level (no redundancy to spare)
# ─────────────────────────────────────────────────────────────────────────────

prune_levels = [0.10, 0.20, 0.30, 0.40, 0.50, 0.60]

def run_pruning_sweep(dataset_name, models_dict, test_loader, fine_tune_loader):
    """
    Progressive pruning sweep across sparsity levels.

    Parameters
    ----------
    models_dict : dict  {display_name: (base_model, fine_tune_epochs)}
        fine_tune_epochs=0  → pretrained baselines (high redundancy, no FT needed)
        fine_tune_epochs=2  → GreenNet (low redundancy, needs recovery)
    """
    results = {name: [] for name in models_dict}
    print(f"\nProgressive Pruning — {dataset_name}")
    print("─" * 60)

    for level in prune_levels:
        print(f"\n  Sparsity {level*100:.0f}%...")
        for name, (base_model, ft_epochs) in models_dict.items():
            set_seed()
            pm = apply_global_pruning(
                base_model,
                amount=level,
                fine_tune_loader=fine_tune_loader if ft_epochs > 0 else None,
                fine_tune_epochs=ft_epochs
            )
            acc, _, _, _, _, _ = evaluate_model(pm, test_loader, device)
            results[name].append(acc)
            ft_note = f" (+{ft_epochs}ep FT)" if ft_epochs > 0 else ""
            print(f"    {name:25s}: acc={acc:.4f}{ft_note}")

    return results


# ── MNIST sweep ───────────────────────────────────────────────────────────────
mnist_models = {
    'MobileNetV2':    (mobilenet_mnist,    0),   # large redundancy — no FT needed
    'ResNet18':       (resnet_mnist,       0),
    'EfficientNet-B0':(efficientnet_mnist, 0),
    'GreenNet':  (custom_mnist,       2),   # compact — 2-epoch FT per level
}

mnist_prune_results = run_pruning_sweep(
    'MNIST', mnist_models, mnist_test_loader, mnist_train_loader)

# ── CIFAR-10 sweep ────────────────────────────────────────────────────────────
cifar_models = {
    'MobileNetV2':    (mobilenet_cifar,    0),
    'ResNet18':       (resnet_cifar,       0),
    'EfficientNet-B0':(efficientnet_cifar, 0),
    'GreenNet':  (custom_cifar,       2),   # compact — 2-epoch FT per level
}

cifar_prune_results = run_pruning_sweep(
    'CIFAR-10', cifar_models, cifar_test_loader, cifar_train_loader)


# ── Plot both side by side ────────────────────────────────────────────────────
colors  = ['#2196F3', '#F44336', '#4CAF50', '#FF9800']
markers = ['o', 's', '^', 'D']
x_vals  = [l * 100 for l in prune_levels]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle('Progressive Pruning: Accuracy vs. Sparsity\n'
             '(Pretrained baselines: no fine-tune | GreenNet: 2-epoch fine-tune per level)',
             fontsize=12, fontweight='bold')

for ax, results, title in [
    (ax1, mnist_prune_results,  'MNIST'),
    (ax2, cifar_prune_results,  'CIFAR-10'),
]:
    for (name, accs), color, marker in zip(results.items(), colors, markers):
        ls = '--' if name == 'GreenNet' else '-'   # dashed = our model
        ax.plot(x_vals, accs, label=name, color=color,
                marker=marker, linewidth=2, markersize=7, linestyle=ls)

    ax.set_xlabel('Pruning Sparsity (%)', fontsize=11)
    ax.set_ylabel('Test Accuracy',        fontsize=11)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)
    ax.set_ylim(0, 1)
    ax.set_xticks([l * 100 for l in prune_levels])

    # Annotate GreenNet values at each point
    ul_accs = results['GreenNet']
    for x, y in zip(x_vals, ul_accs):
        ax.annotate(f'{y:.2f}', (x, y),
                    textcoords='offset points', xytext=(0, 8),
                    fontsize=7, ha='center', color='#FF9800')

plt.tight_layout()
plt.savefig('progressive_pruning.pdf', bbox_inches='tight', dpi=300)
plt.show()
print("✅ Fig 2: Progressive pruning analysis saved.")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 18: Fig 4 — Pareto Frontier: Accuracy vs CO₂ (key contribution)
# The Pareto front shows which models are non-dominated in the
# accuracy–sustainability trade-off space.
# ─────────────────────────────────────────────────────────────────────────────

def pareto_front(df, x_col='CO₂ (kg)', y_col='Accuracy'):
    """Identify Pareto-optimal points (maximize accuracy, minimize CO₂)."""
    pts = df[[x_col, y_col]].values  # [co2, acc]
    is_pareto = np.ones(len(pts), dtype=bool)
    for i, (co2_i, acc_i) in enumerate(pts):
        for j, (co2_j, acc_j) in enumerate(pts):
            if i == j: continue
            if co2_j <= co2_i and acc_j >= acc_i and (co2_j < co2_i or acc_j > acc_i):
                is_pareto[i] = False
                break
    return is_pareto

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

for ax, df, title in [(ax1, mnist_df,  'MNIST'),
                       (ax2, cifar_df, 'CIFAR-10')]:
    is_pareto = pareto_front(df)

    sc = ax.scatter(df['CO₂ (kg)'], df['Accuracy'],
                    s=df['Size (MB)'] * 15 + 50,
                    c=np.log10(df['GMACs'] + 1e-6),
                    cmap='plasma', alpha=0.85, edgecolors='black', linewidth=0.5,
                    zorder=3)

    # Draw Pareto frontier
    pf = df[is_pareto].sort_values('CO₂ (kg)')
    ax.step(pf['CO₂ (kg)'], pf['Accuracy'], where='post',
            color='#E53935', linewidth=2, linestyle='--', label='Pareto Frontier', zorder=4)

    for _, row in df.iterrows():
        ax.annotate(row['Model'],
                    (row['CO₂ (kg)'], row['Accuracy']),
                    textcoords='offset points', xytext=(5, 3),
                    fontsize=7, ha='left')

    ax.set_xscale('log')
    ax.set_xlabel('Training CO₂ Emissions (kg) — log scale', fontsize=10)
    ax.set_ylabel('Test Accuracy', fontsize=10)
    ax.set_title(f'Accuracy–CO₂ Pareto Frontier ({title})\n'
                  'Bubble size ∝ model size, color ∝ log(GMACs)',
                  fontsize=10, fontweight='bold')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.2)
    plt.colorbar(sc, ax=ax, label='log₁₀(GMACs)')

plt.tight_layout()
plt.savefig('pareto_frontier.pdf', bbox_inches='tight', dpi=300)
plt.show()
print("✅ Fig 4: Pareto frontier saved.")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 21: Fig 7 — Green Score Radar Chart (Spider Plot)
# ─────────────────────────────────────────────────────────────────────────────
def radar_chart(df, metrics, title):
    invert = {'CO₂ (kg)', 'Size (MB)', 'Latency (ms)', 'GMACs'}
    n_met  = len(metrics)
    angles = [2 * np.pi * i / n_met for i in range(n_met)] + [0]

    fig, ax = plt.subplots(figsize=(7, 7), subplot_kw={'polar': True})
    colors  = plt.cm.tab10(np.linspace(0, 1, len(df)))

    for (_, row), color in zip(df.iterrows(), colors):
        vals = []
        for m in metrics:
            col_min = df[m].min()
            col_max = df[m].max()
            if col_max == col_min:
                v = 0.5
            else:
                v = (row[m] - col_min) / (col_max - col_min)
                if m in invert:
                    v = 1 - v
            vals.append(v)
        vals += vals[:1]
        ax.plot(angles, vals, color=color, linewidth=2, label=row['Model'])
        ax.fill(angles, vals, color=color, alpha=0.05)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(metrics, size=9)
    ax.set_ylim(0, 1)
    ax.set_yticks([0.25, 0.5, 0.75])
    ax.set_yticklabels(['25%', '50%', '75%'], size=7)
    ax.set_title(title, size=12, fontweight='bold', pad=20)
    ax.legend(loc='upper right', bbox_to_anchor=(1.35, 1.15), fontsize=7)
    ax.grid(alpha=0.3)
    return fig

radar_metrics = ['Accuracy', 'CO₂ (kg)', 'Size (MB)', 'GMACs', 'Latency (ms)', 'Green Score']

fig_m = radar_chart(mnist_df, radar_metrics,
                    'Sustainability Radar — MNIST\n(higher normalized = better)')
fig_m.savefig('radar_mnist.pdf', bbox_inches='tight', dpi=300)
plt.show()                # ← was fig_m.show()

fig_c = radar_chart(cifar_df, radar_metrics,
                    'Sustainability Radar — CIFAR-10\n(higher normalized = better)')
fig_c.savefig('radar_cifar.pdf', bbox_inches='tight', dpi=300)
plt.show()                # ← was fig_c.show()

print("✅ Fig 7: Radar charts saved.")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 22: KD Ablation — Temperature × Alpha on MNIST and CIFAR-10
# ─────────────────────────────────────────────────────────────────────────────

temps  = [1, 2, 4, 8]
alphas = [0.3, 0.5, 0.7, 0.9]

def run_kd_ablation(dataset_name, train_loader, test_loader,
                    teacher, in_channels, base_channels, epochs):
    
    ablation = np.zeros((len(temps), len(alphas)))
    print(f"\nKD Ablation — {dataset_name} "
          f"({epochs} epochs per config, base_channels={base_channels})")
    print("─" * 60)

    for i, T in enumerate(temps):
        for j, alpha in enumerate(alphas):
            set_seed()
            student = GreenNet(
                in_channels=in_channels,
                base_channels=base_channels    # ← critical fix for CIFAR
            ).to(device)

            _, hist = train_with_distillation(
                student, teacher,
                train_loader, test_loader,
                epochs=epochs,
                name=f'Ablation_{dataset_name}_T{T}_a{alpha}',
                device=device,
                temperature=T,
                alpha=alpha)

            acc = hist['val_acc'][-1]
            ablation[i, j] = acc
            print(f"   T={T}, α={alpha:.1f} → acc={acc:.4f}")

    return ablation


# ── Run ablations ─────────────────────────────────────────────────────────────
mnist_ablation = run_kd_ablation(
    dataset_name  = 'MNIST',
    train_loader  = mnist_train_loader,
    test_loader   = mnist_test_loader,
    teacher       = mobilenet_mnist,
    in_channels   = 1,
    base_channels = 16,       # matches Cell 11 MNIST model
    epochs        = 10         # MNIST converges fast
)

cifar_ablation = run_kd_ablation(
    dataset_name  = 'CIFAR-10',
    train_loader  = cifar_train_loader,
    test_loader   = cifar_test_loader,
    teacher       = mobilenet_cifar,
    in_channels   = 3,
    base_channels = 64,       # ← was missing — caused 53% ceiling
    epochs        = 40        # CIFAR needs more steps
)

# ── Plot both heatmaps side by side ───────────────────────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('KD Hyperparameter Ablation: Temperature × Alpha', fontsize=13, fontweight='bold')

for ax, data, title in [
    (ax1, mnist_ablation,  f'MNIST ({10} epochs per config)'),
    (ax2, cifar_ablation,  f'CIFAR-10 ({40} epochs per config)'),
]:
    # Find best cell for annotation
    best_idx = np.unravel_index(np.argmax(data), data.shape)

    sns.heatmap(data,
                xticklabels=[f'α={a}' for a in alphas],
                yticklabels=[f'T={t}' for t in temps],
                annot=True, fmt='.4f', cmap='YlGnBu',
                ax=ax, linewidths=0.5,
                linecolor='white')

    # Highlight best cell with red border
    ax.add_patch(plt.Rectangle(
        (best_idx[1], best_idx[0]), 1, 1,
        fill=False, edgecolor='red', linewidth=3
    ))

    best_T     = temps[best_idx[0]]
    best_alpha = alphas[best_idx[1]]
    best_acc   = data[best_idx]

    ax.set_title(f'{title}\nBest: T={best_T}, α={best_alpha} → {best_acc:.4f}',
                 fontsize=10, fontweight='bold')
    ax.set_xlabel('Alpha (KD soft-loss weight)', fontsize=9)
    ax.set_ylabel('Temperature (T)',             fontsize=9)

plt.tight_layout()
plt.savefig('kd_ablation.pdf', bbox_inches='tight', dpi=300)
plt.show()

# ── Print optimal configs ─────────────────────────────────────────────────────
for name, data in [('MNIST', mnist_ablation), ('CIFAR-10', cifar_ablation)]:
    best_idx   = np.unravel_index(np.argmax(data), data.shape)
    best_T     = temps[best_idx[0]]
    best_alpha = alphas[best_idx[1]]
    best_acc   = data[best_idx]
    print(f"✅ {name} optimal — T={best_T}, α={best_alpha}, acc={best_acc:.4f}")

print("✅ Fig 8: KD ablation heatmap saved.")
````

````python
# ─────────────────────────────────────────────────────────────────────────────
# Cell 26: Final Test Accuracy Evaluation — ALL models, MNIST & CIFAR-10
# ─────────────────────────────────────────────────────────────────────────────

# ── All models with their test loaders ───────────────────────────────────────
mnist_model_registry = [
    ('GreenNet',              custom_mnist,        mnist_test_loader),
    ('GreenNet + KD',         custom_mnist_kd,     mnist_test_loader),
    ('MobileNetV2 (Full)',         mobilenet_mnist,     mnist_test_loader),
    ('MobileNetV2 Pruned 30%',     pruned_mob_mnist,    mnist_test_loader),
    ('ResNet18 Pruned 30%',        pruned_res_mnist,    mnist_test_loader),
    ('EfficientNet-B0 Pruned 30%', pruned_eff_mnist,    mnist_test_loader),
    ('GreenNet INT8',         quant_custom_mnist,  mnist_test_loader),
    ('MobileNetV2 INT8',           quant_mob_mnist,     mnist_test_loader),
]

cifar_model_registry = [
    ('GreenNet',              custom_cifar,        cifar_test_loader),
    ('GreenNet + KD',         custom_cifar_kd,     cifar_test_loader),
    ('MobileNetV2 (Full)',         mobilenet_cifar,     cifar_test_loader),
    ('MobileNetV2 Pruned 30%',     pruned_mob_cifar,    cifar_test_loader),
    ('ResNet18 Pruned 30%',        pruned_res_cifar,    cifar_test_loader),
    ('EfficientNet-B0 Pruned 30%', pruned_eff_cifar,    cifar_test_loader),
    ('GreenNet INT8',         quant_custom_cifar,  cifar_test_loader),
    ('MobileNetV2 INT8',           quant_mob_cifar,     cifar_test_loader),
]


def final_test_evaluation(model_registry, classes, dataset_name):
    """
    Run fresh inference on the test loader for every model.
    Collects predictions, labels, accuracy, precision, ROC-AUC.
    """
    print(f"\n{'='*70}")
    print(f"  RUNNING FRESH TEST INFERENCE — {dataset_name}")
    print(f"{'='*70}")

    fresh_results = []

    for name, model, loader in model_registry:
        # Quantized INT8 models must run on CPU
        is_cpu = 'INT8' in name
        eval_device = torch.device('cpu') if is_cpu else device

        model.to(eval_device)
        acc, prec, loss, roc, preds, labels = evaluate_model(
            model, loader, eval_device)
        model.to(device)    # return to main device after evaluation

        fresh_results.append({
            'Model':    name,
            'Accuracy': acc,
            'Precision':prec,
            'Loss':     loss,
            'ROC-AUC':  roc,
            '_preds':   preds,
            '_labels':  labels,
        })
        print(f"  ✅ {name:<30} Acc={acc:.4f} | Prec={prec:.4f} | ROC={roc:.4f}")

    # ── Summary table ─────────────────────────────────────────────────────────
    print(f"\n{'─'*70}")
    print(f"  {'Model':<30} {'Accuracy':>9} {'Precision':>10} {'ROC-AUC':>9}")
    print(f"  {'─'*30} {'─'*9} {'─'*10} {'─'*9}")
    for r in fresh_results:
        print(f"  {r['Model']:<30} {r['Accuracy']:>9.4f} "
              f"{r['Precision']:>10.4f} {r['ROC-AUC']:>9.4f}")

    print(f"\n{'─'*70}")
    best_acc = max(fresh_results, key=lambda r: r['Accuracy'])
    low_loss = min(fresh_results, key=lambda r: r['Loss'])
    print(f"  🏆 Best accuracy : {best_acc['Model']:<30} {best_acc['Accuracy']:.4f}")
    print(f"  📉 Lowest loss   : {low_loss['Model']:<30} {low_loss['Loss']:.4f}")

    # ── Classification report — ALL models ────────────────────────────────────
    print(f"\n{'='*70}")
    print(f"  CLASSIFICATION REPORTS — {dataset_name}")
    print(f"{'='*70}")

    for r in fresh_results:
        print(f"\n  ── {r['Model']} (Acc={r['Accuracy']:.4f}) "
              f"{'─'*(42-len(r['Model']))}")
        print(classification_report(
            r['_labels'], r['_preds'],
            target_names=classes,
            digits=4,
            zero_division=0
        ))

    # ── Bar chart ─────────────────────────────────────────────────────────────
    fig, ax = plt.subplots(figsize=(13, 4))
    model_names = [r['Model'] for r in fresh_results]
    accs        = [r['Accuracy'] for r in fresh_results]
    colors      = ['#1565C0' if 'GreenNet' in m else '#90A4AE'
                   for m in model_names]

    bars = ax.bar(model_names, accs, color=colors,
                  edgecolor='white', linewidth=0.5)
    ax.set_ylim(min(accs) - 0.05, 1.0)
    ax.set_ylabel('Test Accuracy', fontsize=11)
    ax.set_title(f'Test Accuracy on Test Data — All Models — {dataset_name}\n'
                 f'(blue = our models  |  gray = pretrained baselines)',
                 fontsize=12, fontweight='bold')
    ax.tick_params(axis='x', rotation=30, labelsize=8)
    ax.grid(axis='y', alpha=0.3)

    base_acc = next(r['Accuracy'] for r in fresh_results
                    if r['Model'] == 'GreenNet')
    ax.axhline(y=base_acc, color='#1565C0', linestyle='--', linewidth=1,
               alpha=0.5, label=f'GreenNet baseline ({base_acc:.4f})')
    ax.legend(fontsize=8)

    for bar, v in zip(bars, accs):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.003,
                f'{v:.4f}', ha='center', fontsize=7.5, fontweight='bold')

    plt.tight_layout()
    fname = f"test_accuracy_{dataset_name.lower().replace('-','').replace(' ','')}.pdf"
    plt.savefig(fname, bbox_inches='tight', dpi=300)
    plt.show()
    print(f"✅ Chart saved → {fname}")


# ── Run for both datasets ─────────────────────────────────────────────────────
final_test_evaluation(mnist_model_registry,  MNIST_CLASSES,  'MNIST')
final_test_evaluation(cifar_model_registry,  CIFAR_CLASSES,  'CIFAR-10')
````
